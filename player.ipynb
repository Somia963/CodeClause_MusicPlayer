```python
from tkinter import Tk, Listbox, Button
from pygame import mixer
from tkinter import filedialog

# Function to add songs to the playlist
def addsongs():
 
    song_paths = filedialog.askopenfilenames()

    
    for song_path in song_paths:
        
        song_name = song_path.split("/")[-1]  # Extract the last part of the path

      
        playlist.insert("end", song_name)

# Function to delete song from the playlist
def deletesong():
    # Get the index of the selected song in the playlist
    selected_index = playlist.curselection()

    # Ensure song is selected
    if not selected_index:
        return

    # Remove  from the playlist
    playlist.delete(selected_index)


def Play():
    
    selected_index = playlist.curselection()

    # Ensure that a song is selected
    if not selected_index:
        return

    # Get the selected song from the playlist
    selected_song = playlist.get(selected_index)

    
    mixer.music.stop()

    # Load 
    mixer.music.load(selected_song)
    mixer.music.play()

# Function to play the next 
def Next():
    
    selected_index = playlist.curselection()


    if not selected_index:
        return
    total_songs = playlist.size()

    # Increment the current song index to move to the next song
    next_index = (selected_index[0] + 1) % total_songs

    # Select the next song in the playlist
    playlist.selection_clear(0, "end")
    playlist.selection_set(next_index)

    # Play the next song
    Play()


def Previous():
 
    selected_index = playlist.curselection()

    if not selected_index:
        return

    # Get  the playlist
    total_songs = playlist.size()

    # Decrement the current song index to move to the previous song
    prev_index = (selected_index[0] - 1) % total_songs

    # Select the previous song in the playlist
    playlist.selection_clear(0, "end")
    playlist.selection_set(prev_index)

    # Play the previous song
    Play()


def Resume():
    mixer.music.unpause()

def Stop():
    mixer.music.stop()


def Pause():
    mixer.music.pause()

# Create the main window
root = Tk()
root.geometry("230x200")
root.configure(bg='purple')
root.title("Python MP3 Player")

# Initialize the mixer
mixer.init()


playlist = Listbox(root, selectmode='single', bg='white')
playlist.grid(row=0, column=0, columnspan=9)


play_button = Button(root, text="Play", command=Play)
play_button.grid(row=1, column=0)

next_button = Button(root, text="Next", command=Next)
next_button.grid(row=1, column=1)

previous_button = Button(root, text="Previous", command=Previous)
previous_button.grid(row=1, column=2)

resume_button = Button(root, text="Resume", command=Resume)
resume_button.grid(row=1, column=3)

stop_button = Button(root, text="Stop", command=Stop)
stop_button.grid(row=1, column=4)

pause_button = Button(root, text="Pause", command=Pause)
pause_button.grid(row=1, column=5)

add_songs_button = Button(root, text="Add Songs", command=addsongs)
add_songs_button.grid(row=2, column=0)

delete_song_button = Button(root, text="Delete Song", command=deletesong)
delete_song_button.grid(row=2, column=1)


root.mainloop()
```
