```python
import tkinter as tk
from tkinter import messagebox, filedialog
import pygame


def play_music():
    pygame.mixer.music.load(filename)
    pygame.mixer.music.play()


def next_music():
    pygame.mixer.music.stop()
    messagebox.showinfo('Next', 'Next button clicked')


def prev_music():
    pygame.mixer.music.stop()
    messagebox.showinfo('Previous', 'Previous button clicked')


def stop_music():
    pygame.mixer.music.stop()

# Function to handle the 'Resume' button click event
def resume_music():
    pygame.mixer.music.unpause()

# Function to handle the 'Open' menu item click event
def open_file():
    global filename
    filename = filedialog.askopenfilename()
    messagebox.showinfo('Open', f'Selected file: {filename}')

#  Tkinter & pygame
root = tk.Tk()
root.title('Music Player')
root.geometry("230x200")
root.configure(bg='purple')

pygame.mixer.init()

# Create menu
menu_bar = tk.Menu(root)
root.config(menu=menu_bar)

file_menu = tk.Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label='File', menu=file_menu)
file_menu.add_command(label='Open', command=open_file)
file_menu.add_separator()
file_menu.add_command(label='Exit', command=root.quit)

# control
control_frame = tk.Frame(root, bg='purple')
control_frame.pack()


play_btn = tk.Button(control_frame, text='Play', command=play_music, bg='white')
play_btn.grid(row=0, column=0)

prev_btn = tk.Button(control_frame, text='Previous', command=prev_music, bg='white')
prev_btn.grid(row=0, column=1)

next_btn = tk.Button(control_frame, text='Next', command=next_music, bg='white')
next_btn.grid(row=0, column=2)

stop_btn = tk.Button(control_frame, text='Stop', command=stop_music, bg='white')
stop_btn.grid(row=0, column=3)

resume_btn = tk.Button(control_frame, text='Resume', command=resume_music, bg='white')
resume_btn.grid(row=0, column=4)

root.mainloop()
```
